Estimate difference between observation date and production datetime for the year 2024

```python
import glob
from products.filenames import timestamp_viirs_to_datetime
from pathlib import Path
from datetime import datetime,timedelta
import numpy as np
import pandas as pd

vnp10a1_folder = '/home/imperatoren/work/VIIRS_S2_comparison/data/V10A1/VNP10A1'
viirs_l3_filepats_2024 = glob.glob(vnp10a1_folder+'/*A2024*.h5')
```

```python
obervation_production_differences = []
for filepath in viirs_l3_filepats_2024:
    _, observation_day_str, _, _, production_datetime_str, _, = Path(filepath).name.split(".")
    observation_day = timestamp_viirs_to_datetime(observation_day_str[1:])
    end_of_observation_day_datetime = observation_day + timedelta(days=1)
    production_datetime = datetime.strptime(production_datetime_str, f"%Y%j%H%M%S")
    obervation_production_difference = production_datetime - end_of_observation_day_datetime 
    obervation_production_differences.append(obervation_production_difference)

obervation_production_differences = pd.to_timedelta(pd.Series(obervation_production_differences))
mean_difference = obervation_production_differences.mean()
print("The mean difference is ", mean_difference.days, "days and ", mean_difference.seconds / 3600, " hours after the midnight after observation")
median_difference = obervation_production_differences.median()
print("The median difference is ", median_difference.days, "days and ", median_difference.seconds / 3600, " hours after the midnight after observation")
min_difference = obervation_production_differences.min()
print("The min difference is ", min_difference.days, "days and ", min_difference.seconds / 3600, " hours after the midnight after observation")
max_difference = obervation_production_differences.max()
print("The max difference is ", max_difference.days, "days and ", max_difference.seconds / 3600, " hours after the midnight after observation")
```
